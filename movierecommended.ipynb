```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import difflib
```

```python
df=pd.read_csv('movies1.csv')
```

```python
df.head()
```

```python
df=df[['index','genres','keywords','tagline','title','cast','director',]]
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
selected_features=['genres','keywords','tagline','title','cast','director']

for feature in selected_features:
    df[feature]=df[feature].fillna("")
```

```python
df.isnull().sum()
```

```python

def get_important_features(data):
    important_features=[]
    for i in range(0,data.shape[0]):
        important_features.append(df['genres'][i]+' '+df['keywords'][i]+' '+df['tagline'][i]+' '+df['title'][i]+' '+df['cast'][i]+' '+df['director'][i])

    return important_features

df['important_features']=get_important_features(df)
```

```python
df.head()
```

```python
cm=CountVectorizer().fit_transform(df['important_features'])

cs=cosine_similarity(cm)
```

```python
cm
```

```python
cs
```

```python
cs.shape
```

```python
movie_name=input("enter favourite movie name : ")
```

```python

get_list_allmovies=df['title'].tolist()
```

```python
get_list_allmovies
```

```python
find_close_match=difflib.get_close_matches(movie_name,get_list_allmovies)
```

```python
find_close_match
```

```python
closest_match=find_close_match[0]

index_of_the_movie=df[df.title==closest_match]['index'].values[0]
```

```python
index_of_the_movie
```

```python
similarity_score=list(enumerate(cs[index_of_the_movie]))


sorted_similarity_scores=sorted(similarity_score ,key= lambda x : x[1] ,reverse=True)
```

```python
sorted_similarity_scores
```

```python
i=1
print('similar movies are :')
for movie in sorted_similarity_scores:
    index=movie[0]

    similar_movie_name=df[df.index==index]['title'].values[0]
    if i<=10:
        print(i,' : ',similar_movie_name)
        i=i+1
    else:
        break
```
