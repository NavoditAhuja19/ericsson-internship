```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
df = pd.read_csv("Mobile.csv")
 
```

```python
df.head()
```

```python
X = df.iloc[:,0:20]  
y = df.iloc[:,-1] 
```

```python
X.head()
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
final = bestfeatures.fit(X,y)
```

```python
dfscores = pd.DataFrame(final.scores_)
dfcolumns = pd.DataFrame(X.columns)
```

```python
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Scores'] 
```

```python
featureScores
```

```python
temp_df=featureScores.nlargest(10,'Scores')
```

```python
size=temp_df.shape
```

```python
imp_columns=[]
for i in range(0,size[0]):
    imp_columns.append(temp_df.iloc[i,0])
```

```python
important_feature_df=df[imp_columns]
```
